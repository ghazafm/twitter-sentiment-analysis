```python
import pandas as pd 
df = pd.read_csv('D:\BCC\Projek_Intern\SocialMediaSentiment\preprocessing\EDA\datasets\gabungan.csv')
```

```python

# Hitung jumlah tweet untuk setiap user
user_tweet_counts = df['User ID'].value_counts()

# Gabungkan jumlah tweet ke DataFrame awal
df_with_counts = df.merge(user_tweet_counts.rename('Count'), left_on='User ID', right_index=True)

# Urutkan berdasarkan jumlah tweet
sorted_df = df_with_counts.sort_values(by='Tweet Count', ascending=False)

# Ambil 10 user teratas beserta tweetnya
top_10_users_with_tweets = pd.DataFrame(sorted_df.groupby('User ID').head(10))

# Tampilkan hasil
top_10_users_with_tweets
```

```python

df_pos = df[df['label']=='Positive']
df_positive = df[df['label']=='Positive']
df_positive = df_positive.groupby('User ID').size().nlargest(10).reset_index(name='Count')
df_positive = pd.DataFrame(df_positive)
df_positive = pd.merge(df_positive, df_pos, on='User ID')
df_positive.to_csv('positive.csv')
df_positive

# user_positive = positive_df['User ID'].value_counts()

# df_positive = df.merge(user_positive.rename('Count'), left_on='User ID', right_index=True)

# df_positive = df_positive.sort_values(by='Count', ascending=False)

```
